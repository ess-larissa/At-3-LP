# ***Projeto vendas***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
vendas = pd.read_csv('/content/drive/MyDrive/vendas_tratado.csv' , sep=',')
```

```python
vendas
```

# ***Criação da série***

```python
vendas2 = vendas.drop(columns='Ano')
vendas2.head()
```

```python
# Transformar em Array??
vendas3 = vendas2.values
vendas3
```

```python
# Transformar em lista
vendas4 = list(vendas3.flatten())
print(vendas4)
```

```python
vendas4 = vendas3.flatten().astype(float).tolist()
print(vendas4)
```

```python
indice = pd.date_range ('2000', periods = len(vendas4), freq = 'M')
indice
```

```python
serie = pd.Series(vendas4, index = indice)
serie
```

```python
serie.plot()
plt.show()
```

# ***Análise e preparação***

# ***Média móvel***

```python
media_movel = serie.rolling(window=6)
media_movel = media_movel.mean()
```

```python
plt.plot(media_movel)
```

```python
plt.plot(serie, label='Série Real')
plt.plot(media_movel,color='red', label='Média Móvel 6 meses')
plt.legend(loc='best')
plt.show()
```

# ***Decomposição***

Análise de tendência, sazonalidae e resíduos.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposicao = seasonal_decompose(serie)
```

```python
decomposicao.plot()
```

# ***Normalidade e transformação***

```python
import scipy.stats as stats
```

```python
stats.probplot(serie, dist='norm', plot=plt)
plt.title('Normal QQ plot')
plt.show()
```

# ***Teste Shapiro-Wilk***

Critérios:

Nível de significância de 0,05 ou 5%

Ho = distribuição normal p > 0,05

Ha = distribuição não normal p <= 0,05

```python
e, p = stats.shapiro(serie)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

Distribuição não normal

# ***Diferenciação***

```python
serie_diff = serie.diff().dropna()
serie_diff
```

# ***Histograma***

```python
import seaborn as sns
sns.histplot(serie, kde=True);
```

```python
sns.histplot(serie_diff, kde=True)
```

```python
stats.probplot(serie_diff, dist='norm', plot=plt)
plt.title('Normal QQ plot')
plt.show()
```

```python
e, p = stats.shapiro(serie_diff)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

# ***Estacionaridade***

```python
import statsmodels.tsa.stattools
```

Teste KPSS (Kwiatkowski-Philliips-Schmidt-Shin)

Ha = não é estacionário: estatística do teste > valor crítico

Ho = é estacionário estatística do teste < valor crítico

valor crítico = 5%

```python
kpss = statsmodels.tsa.stattools.kpss(serie_diff)
print ('Estatística do teste: {:.4f}' .format(kpss[0]))
print ('p_valor: {:.4f}' .format(kpss[1]))
print ('Valores críticos:')
for chave, valor in kpss[3].items():
  print ('{}: {:.3f}' .format(chave, valor))
```

É estacionário

# ***Autocorrelação***

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(serie_diff, lags=20)
plt.show()
```

```python
plot_pacf(serie_diff, lags=20)
plt.show()
```

Série autocorrelacionada, não tem problema, pois é um pressuposto da análise dos resíduos dos modelos das séries temporais.

# ***ARIMA***

# ***Modelo AR***

Modelo ARIMA: (p, d, q)

Modelo AR: (p, 0, 0)


```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
modelo_ar = ARIMA(serie_diff, order=(10,0,0))
```

```python
resultado = modelo_ar.fit()
print(resultado.summary())
```

Usar AIC ou BIC

AIC       :                    1914.145 (10, 0, 0)


```python
plot_acf(serie_diff, lags=20)
plt.show()
```

# ***Análine de resíduos***

Problemas de normalidade e de autocorrelação

```python
residuos = resultado.resid
```

```python
residuos.plot()
plt.show
```

# ***Normalidade***

```python
stats.probplot(residuos, dist='norm', plot=plt)
plt.title('Normal QQ plot')
plt.show()
```

```python
e, p = stats.shapiro(residuos)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

Distribuição não é normal

```python
sns.distplot(residuos, kde=True)
```

# ***Autocorrelação***

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(residuos, lags=20)
plt.show()
```

```python
plot_pacf(residuos, lags=20)
plt.show()
```

Existe autocorrelação entre os resíduos

```python
plt.plot(serie_diff, label='Série Real')
plt.plot(serie_diff-residuos,color='red', label='Resíduos')
plt.legend(loc='best')
plt.show()
```

# ***Modelo ARMA***

Modelo ARIMA: (p,d,q)

Modelo ARMA: (p,0,q)

Modelo AR: AIC =  1914.145 (10, 0, 0)

Modelo MA: AIC = 1825.098 (0,0,10)

Modelo ARMA: AIC = 1561.930 (10, 0, 10)

```python
residuos_arma.plot()
plt.show
```

# ***Normalidade***

```python
stats.probplot(residuos_arma, dist='norm', plot=plt)
plt.title('Normal QQ plot')
plt.show()
```

```python
e, p = stats.shapiro(residuos_arma)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

Distribuição dos resíduos não normal - aprox. normal

```python
import seaborn as sns
sns.distplot(residuos_arma, kde=True)
```

# ***Autocorrelação***

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(residuos_arma, lags=20)
plt.show()
```

resíduos autocorrelacionados

Esse modelo consegue prever prazos maiores, significa que é um modelo melhor

# ***ARIMA***

Modelo ARIMA: (p, d, q)

Modelo AR: AIC = 1914.145 (10, 0, 0)

Modelo MA: AIC = 1825.098 (0,0,10)

Modelo ARMA: AIC = 1561.930 (10, 0, 10)

Modelo ARIMA: AIC = 1621.850 (10, 1, 10)

# ***Análise de resíduos***

```python
residuo_arima.plot()
plt.show()
```

# ***Normalidade***

```python
e, p = stats.shapiro(residuo_arima)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

Distribuição não é normal

```python
import seaborn as sns
sns.distplot(residuo_arima, kde=True)
```

# ***Autocorrelação***

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(residuo_arima, lags=20)
plt.show()
```

Valores fora do intervalo de confiança, mostrando que os resíduos estão autocorrelacionados
